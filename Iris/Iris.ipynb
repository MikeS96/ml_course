```python
# Importing relevant libraries
import cv2
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import mglearn

%matplotlib inline
```

### Load the Iris Dataset

```python
iris_dataset = pd.read_csv('dataset/Iris.csv')
iris_dataset
```

## Visualizing head

```python
iris_dataset.head()
```

### Show the DF keys

```python
print('Target names: {}\n'.format(np.unique(iris_dataset['Species'])))
print('Keys of Iris dataset: {}\n'.format(iris_dataset.keys()))
print('Feature names: {}\n'.format(iris_dataset.keys()[1:5]))
```

### Extract Features and Targets

```python
# Features
#X = iris_dataset.drop(['Id', 'Species'], axis=1).to_numpy()
X = iris_dataset.drop(['Id', 'Species'], axis=1)
print("Features shape: {0}".format(X.shape))
print("First sample of the feature matrix: \n{0}\n".format(X.loc[0:4, :]))
# Targets
#Y = iris_dataset[['Species']].to_numpy()
Y = iris_dataset['Species']
print("Targets shape: {0}".format(Y.shape))
print("First sample of the target vector: \n{0}".format(Y.loc[0:4]))
```

## Factorizing the class's labels

```python
Y, class_l = pd.factorize(Y)
print("The Factorized labeld are: \n{0}".format(Y))
print("Class labels: \n{0}".format(class_l))
```

### Splitting data into train and test

```python
# Train test split
X_train, X_test, y_train, y_test = train_test_split(X, Y, stratify = Y,  random_state = 140)
```

```python
print('Training feature matrix size: \n{}'.format(X_train.shape))
print('Test feature matrix size: \n{}'.format(X_test.shape))
print('Training target vector size: \n{}'.format(y_train.shape))
print('Test target vector size: \n{}'.format(y_test.shape))
print('The targets for training are: \n{}'.format(y_train))
```

## Visualizing data with scatter matrix

The next plot presents the scatter matrix of the features presented in the Iris Dataset. In fact, it can be seen that the classes Iris-versicolor and Iris-virginica share a similar feature space (as for almost all the plots these classes are very close) whereas the Iris-setosa is always away from the two classes previously mentioned. That means that for this particular classification task, the algoritm will easily identify the class Iris-Setosa but will probably struggle a little more differencing between Iris-versicolor and Iris-virginica.

```python
# Label the columns using the string in iris_dataset
iris_dataframe_train = pd.DataFrame(X_train, columns = X.keys())
# Create a scatter matrix from the dataframe, color by y_train
pd.plotting.scatter_matrix(iris_dataframe_train, c = y_train, figsize = (15, 15), marker = 'o',
                          hist_kwds = {'bins': 20}, s = 60, alpha = .8, cmap = mglearn.cm3)
# Plot labels
handles = [plt.plot([],[],color=plt.cm.brg(i/2.), ls="", marker=".", \
                    markersize=np.sqrt(10))[0] for i in range(3)]
labels = class_l
plt.legend(handles, labels, loc=(1.02,0))
plt.show()
```

## Complexity analysis

The following lines of code aim to reproduce a complexity analysis using the KNN model changing the number of neighbors' parameter.

```python
from sklearn.neighbors import KNeighborsClassifier
# Lists to store accuracy over trainings
training_accuracy = []
test_accuracy = []
# try neighbors_settings from 1 to 10
neighbors_settings = range(1, 20)

max_test_acc = 0
best_neighs = 0

for n_neighbors_ in neighbors_settings:
    # Build the model
    clf = KNeighborsClassifier(n_neighbors = n_neighbors_)
    clf.fit(X_train, y_train)
    # Record training set accuracy
    training_accuracy.append(clf.score(X_train, y_train))
    # Record generalization accuracy
    test_acc = clf.score(X_test, y_test)
    test_accuracy.append(test_acc)
    # Storing best number of neighbords
    if max_test_acc <= test_acc:
        max_test_acc = test_acc
        best_neighs = n_neighbors_
    
plt.plot(neighbors_settings, training_accuracy, label = 'Training accuracy')
plt.plot(neighbors_settings, test_accuracy, label = 'Test accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Number of Neighbors')
plt.legend()

print('The maximum training accuracy obtained was: {}'.format(max(training_accuracy)))
print('The maximum test accuracy obtained was {0} with {1} neighbords'.format(max_test_acc, best_neighs))
```

From the plot produced above, we can identify that the model produces the best results using 3, 11 and 13 neighbors respectively. That is said due to with those configurations, the model is capable to obtain the highest test accuracy. However, as the complexity of the KNN model is inverse to the number of neighbors, the simplest model would be the one with 13 neighbors, therefore, from the complexity analysis I would say that the chosen model is **KNN with 13 neighbors**.

## Plotting confusion matrix

Confusion matrix analysis for the model chosen from the complexity analysis.

```python
# Make a prediction
from sklearn.metrics import confusion_matrix

clf = KNeighborsClassifier(n_neighbors = best_neighs)
clf.fit(X_train, y_train)
y_pred_test = clf.predict(X_test)

cm = confusion_matrix(y_test, y_pred_test)
print('The confusion matrix of the model is: \n{}'.format(cm))
```

```python
from sklearn.metrics import plot_confusion_matrix

plot_confusion_matrix(clf, X_test, y_test, display_labels = labels)
```

```python
# Computing accuracy
print('Test set score: {:.2f}'.format(clf.score(X_test, y_test)))
```

The results produced by the confusion matrix show that the model has a really high test accuracy (97%). The model only has 3 errors (False positive) with the class label Iris-versicolor, when the true label was actually Iris-Virginica. In a similar fashion, the predicted label Iris-Virginica has 3 False negatives, as the model did not classify 3 samples of Iris-Virginica in a correct way. 

With this information we can conclude that the KNN model with 13 neighbors is a good model selection for this particular classification problem due to the simplicity of the model. This model is simple as the complexity of the model is inverse to the number of neighbors, therefore, for more neighbors, less complexity for the KNN model.

## Visualizing model complexity (KNN) in a simpler dataset

```python
X, y = mglearn.datasets.make_forge()
fig, axes = plt.subplots(1, 3, figsize = (10, 3))
for n_neighbors, ax in zip([1, 3, 9], axes):
    # The fit method returns the object self, so we can instantiate
    # and fir in one line
    clf = KNeighborsClassifier(n_neighbors = n_neighbors).fit(X, y)
    mglearn.plots.plot_2d_separator(clf, X, fill = True, eps = 0.5, ax = ax, alpha = .4)
    mglearn.discrete_scatter(X[:, 0], X[:, 1], y, ax = ax)
    ax.set_title("{} neighbor(s)".format(n_neighbors))
    ax.set_xlabel("Feature 0")
    ax.set_ylabel("Feature 1")
axes[0].legend(loc = 3)
```
